```python
mean=np.mean(ucs420_manya)
print("mean:",mean)
```

```python
median=np.median(ucs420_manya)
print("median:",median)
```

```python
max=np.max(ucs420_manya)
print("max:",max)
```

```python
min=np.min(ucs420_manya)
print("min:",min)
```

```python
unique=np.unique(ucs420_manya)
print("unique:",unique)
```
