```python
#4
import numpy as np
manya=np.linspace(10,100,25)
print("original array :\n",manya)
print("dimensions:",manya.ndim)
print("shape:",manya.shape)
print("size:",manya.size)
print("data type :",manya.dtype)
print("bytes consumed by array:",manya.nbytes)
```

```python
transpose_arr=manya.reshape(25,1)
print("transpose of arr using reshape\n",transpose_arr)
#cannot use T attribute for 1Darray as transpose only applicable for multidimensional arrays
```
