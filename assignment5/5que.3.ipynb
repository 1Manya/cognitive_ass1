```python
import numpy as np
initials= ('A','B')
X = sum(ord(char) for char in initials)
sales= np.array([X,X+50,X+100,X+150,X+200])
print("sales data",sales)
tax_rate=((X%5)+5)/100
taxes = sales*tax_rate
print("taxes applied :",taxes)
discounted_sales = np.where(sales <X+100, sales * 0.95, sales*0.90)
print("discounted sales :",discounted_sales)
week=3
sales_matrix= np.tile(discounted_sales,(week,1))
print("sales matrix :",sales_matrix)

#apply 2% weekly increase
weekly_multiplier=np.array([1.00,1.02,1.04]).reshape(week,1)
updated_sales_matrix= sales_matrix*weekly_multiplier
print("updated sales matrix :",updated_sales_matrix)
```
