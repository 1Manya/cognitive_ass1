```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-10,10,100)
y1=x**2
y2=np.sin(x)
y3=np.exp(x)
y4=np.log(np.abs(x)+1)
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.plot(x,y1,label=r'y=x^2$')
plt.title('quadratic function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.subplot(2,2,2)
plt.plot(x,y2,label=r'y=$\sin(x)$')
plt.title('sine function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.subplot(2,2,3)
plt.plot(x,y3,label=r'y=e^x')
plt.title('exponential function')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(0,100)
plt.grid(True)
plt.legend()

plt.subplot(2,2,4)
plt.plot(x,y4,label=r'y=ln(x)')
plt.title('logarithmic function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
