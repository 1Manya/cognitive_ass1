```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = int(input("Enter your roll number as seed: "))
np.random.seed(roll_number)
data = np.random.randn(50)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(np.cumsum(data), color='b')
axs[0, 0].set_title('Cumulative Sum')
axs[0, 0].grid(True)

axs[0, 1].scatter(range(50), data, color='r', alpha=0.6)
axs[0, 1].set_title('Scatter Plot with Noise')
axs[0, 1].grid(True)

plt.tight_layout()
plt.show()


```
