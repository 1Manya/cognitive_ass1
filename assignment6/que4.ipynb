```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
df = pd.read_csv(url)

# 1. Line plot for total profit
plt.figure(figsize=(8, 5))
sns.lineplot(x=df['month_number'], y=df['total_profit'], marker='o', label="Total Profit")
plt.title('Total Profit Over Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.legend()
plt.show()

# 2. Multiline plot for product sales
plt.figure(figsize=(10, 6))
for column in df.columns[1:-1]:
     sns.lineplot(x=df['month_number'], y=df[column], label=[column])

plt.title('Product Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

# 3. Bar chart for all features
df.set_index('month_number').plot(kind='bar', figsize=(12, 6))
plt.title('Monthly Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.show()


```
