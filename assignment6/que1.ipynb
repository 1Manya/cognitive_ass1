```python
import numpy as np
import matplotlib.pyplot as plt
M=float(input("enter a value for M:"))
x=np.linspace(-10,10,100)
y1=M*x**2
y2=M*np.sin(x)
plt.figure(figsize=(8,5))
plt.plot(x,y1,label="$y=mx^2$",color="b",linestyle='-')
plt.plot(x,y2,label="$y=mxsin(x)$",color="r",linestyle='--')
plt.title("Mathematical function plots ")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
