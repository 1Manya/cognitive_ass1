```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data={'subjects':['math','science','english','history','art'],'score':[85,90,78,88,76]}
df=pd.DataFrame(data)
print(df)
plt.figure(figsize=(8,5))
ax=sns.barplot(x='subjects',y='score',data=df)
for i,score in enumerate(df['score']):
    ax.text(i,score+1,str(score),ha='center')
plt.title('Subject Scores')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.grid(axis='y')
plt.show()
```
